```python
import os
import pandas as pd

# Set the directory where your files are stored
directory = './yahoo_/yahoo_preprocessed/'  # Replace with the path to your files

# Initialize an empty list to store the individual DataFrames
data_frames = []

# Loop through all the files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):  # Assuming the files are in CSV format
        # Construct the full file path
        file_path = os.path.join(directory, filename)
        
        # Read the file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Extract the ticker symbol from the filename (assuming it's the first part of the file name before an underscore)
        ticker = filename.split('_')[0]
        
        # Add the 'Ticker' column to the DataFrame
        df['Ticker'] = ticker
        
        # Append the DataFrame to the list
        data_frames.append(df)

# Concatenate all the DataFrames in the list into a single DataFrame
combined_df = pd.concat(data_frames, ignore_index=True)

# Save the combined DataFrame to a new CSV file (optional)
combined_df.to_csv('./yahoo_/combined_data/combined_stock_data.csv', index=False)

# Now combined_df is your consolidated DataFrame with all the stock data
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier, GradientBoostingRegressor
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestRegressor


# Load your data into a pandas DataFrame
df = pd.read_csv('./yahoo_/yahoo_preprocessed/AAL_preprocessed.csv')

# Feature Engineering: Add rolling averages and momentum indicators
df['SMA_10'] = df['Close'].rolling(window=10).mean()
df['Momentum_5'] = df['Close'] - df['Close'].shift(5)

# Target Variables
df['Target'] = df['Close'].shift(-1)  # Next day's closing price
df['Direction'] = (df['Target'] > df['Close']).astype(int)  # 1 if price goes up, 0 otherwise
df['Magnitude'] = (df['Target'] - df['Close']) / df['Close']  # Percentage change
df.dropna(inplace=True)  # Drop rows with NaN values

# Split the data into features and labels
X = df.drop(['Date', 'Close', 'Adj Close', 'Target', 'Direction', 'Magnitude'], axis=1)
y_class = df['Direction']
y_reg = df['Magnitude']

# Train/Test Split without shuffling
X_train, X_test, y_class_train, y_class_test, y_reg_train, y_reg_test = train_test_split(
    X, y_class, y_reg, test_size=0.2, random_state=42, shuffle=False
)

# Preprocessing pipelines for both numeric and categorical data
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = ['Day_of_week']

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

# Classification Model Pipeline
classification_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('feature_selection', SelectFromModel(RandomForestClassifier(n_estimators=100))),
    ('classifier', RandomForestClassifier(random_state=42))])

# Regression Model Pipeline
regression_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('feature_selection', SelectFromModel(GradientBoostingRegressor(n_estimators=100))),
    ('regressor', GradientBoostingRegressor(random_state=42))])

# Define separate parameter grids
param_grid_class = {
    'classifier__n_estimators': [100, 200],
    'classifier__max_depth': [5, 10]
}

param_grid_reg = {
    'regressor__n_estimators': [100, 200],
    'regressor__learning_rate': [0.01, 0.1]
}

# Separate grid searches
grid_search_class = GridSearchCV(classification_pipeline, param_grid=param_grid_class, cv=5, n_jobs=-1)
grid_search_reg = GridSearchCV(regression_pipeline, param_grid=param_grid_reg, cv=5, n_jobs=-1)

# Model Training and Evaluation
# Classification
grid_search_class.fit(X_train, y_class_train)
y_class_pred = grid_search_class.best_estimator_.predict(X_test)
class_accuracy = accuracy_score(y_class_test, y_class_pred)

# Regression
grid_search_reg.fit(X_train, y_reg_train)
y_reg_pred = grid_search_reg.best_estimator_.predict(X_test)
reg_rmse = mean_squared_error(y_reg_test, y_reg_pred, squared=False)

print(f"Classification Accuracy: {class_accuracy}")
print(f"Regression RMSE: {reg_rmse}")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, RandomizedSearchCV, TimeSeriesSplit
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier, GradientBoostingRegressor, StackingRegressor
from sklearn.linear_model import Ridge
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectFromModel
from scipy.stats import randint, uniform
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestRegressor


# Load your data into a pandas DataFrame
df = pd.read_csv('./yahoo_/yahoo_preprocessed/AAL_preprocessed.csv')

# Feature Engineering: Add rolling averages, momentum indicators, and interaction features
df['SMA_10'] = df['Close'].rolling(window=10).mean()
df['Momentum_5'] = df['Close'] - df['Close'].shift(5)
df['SMA_20'] = df['Close'].rolling(window=20).mean()
df['SMA_50'] = df['Close'].rolling(window=50).mean()
df['Momentum_10'] = df['Close'] - df['Close'].shift(10)
df['SMA_20_cross_SMA_50'] = df['SMA_20'] - df['SMA_50']  # Interaction feature

# Add more lag features
for lag in range(1, 6):
    df[f'Lag_{lag}'] = df['Close'].shift(lag)

# Drop the NaN values created by lag features
df.dropna(inplace=True)

# Define your target variables
df['Target'] = df['Close'].shift(-1)  # Next day's closing price
df['Direction'] = (df['Target'] > df['Close']).astype(int)  # 1 if price goes up, 0 otherwise
df['Magnitude'] = (df['Target'] - df['Close']) / df['Close']  # Percentage change

# Split the data into features and labels
X = df.drop(['Date', 'Close', 'Adj Close', 'Target', 'Direction', 'Magnitude'], axis=1)
y_class = df['Direction']
y_reg = df['Magnitude']

# Train/Test Split without shuffling
X_train, X_test, y_class_train, y_class_test, y_reg_train, y_reg_test = train_test_split(
    X, y_class, y_reg, test_size=0.2, random_state=42, shuffle=False
)

# Assuming y_reg_test is a pandas Series
mask = y_reg_test.notna()  # Create a mask for non-NaN values
y_reg_test_clean = y_reg_test[mask]  # Apply the mask to y_reg_test
y_class_test_clean = y_class_test[mask]  # Apply the same mask to y_class_test to keep the alignment
X_test_clean = X_test.loc[mask]  # Apply the same mask to X_test to keep the alignment

# After filtering NaN values, reset the index
X_test_clean.reset_index(drop=True, inplace=True)
y_class_test_clean.reset_index(drop=True, inplace=True)
y_reg_test_clean.reset_index(drop=True, inplace=True)

# Preprocessing pipelines for both numeric and categorical data
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()),
    ('pca', PCA(n_components=0.95))  # Apply PCA to reduce dimensionality
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
])

# Create a stacking regressor as an example for the regression model
final_estimator = GradientBoostingRegressor(n_estimators=100, random_state=42)
estimators = [
    ('rf', RandomForestRegressor(n_estimators=100, random_state=42)),  # Changed to regressor
    ('ridge', Ridge(random_state=42))
]
stacking_regressor = StackingRegressor(
    estimators=estimators,
    final_estimator=final_estimator
)


# Classification Model Pipeline
classification_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('feature_selection', SelectFromModel(RandomForestClassifier(n_estimators=100))),
    ('classifier', RandomForestClassifier(random_state=42))
])

# Regression Model Pipeline
regression_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('feature_selection', SelectFromModel(GradientBoostingRegressor(n_estimators=100))),
    ('regressor', stacking_regressor)
])

# Define parameter grids as before, consider expanding with more options
# Define parameter distributions
param_dist_class = {
    'feature_selection__estimator__n_estimators': randint(100, 500),
    'feature_selection__estimator__max_depth': randint(3, 20),
    'feature_selection__estimator__min_samples_split': uniform(0.01, 0.1),
    'feature_selection__estimator__min_samples_leaf': uniform(0.01, 0.1),
    'classifier__n_estimators': randint(100, 500),
    'classifier__max_depth': randint(3, 20),
    'classifier__min_samples_split': uniform(0.01, 0.1),
    'classifier__min_samples_leaf': uniform(0.01, 0.1)
}

param_dist_reg = {
    'feature_selection__estimator__n_estimators': randint(100, 500),
    'feature_selection__estimator__max_depth': randint(3, 20),
    'feature_selection__estimator__learning_rate': uniform(0.01, 0.1),
    'regressor__final_estimator__n_estimators': randint(100, 500),
    'regressor__final_estimator__max_depth': randint(3, 10),
    'regressor__final_estimator__learning_rate': uniform(0.01, 0.1)
}

# RandomizedSearchCV instead of GridSearchCV for efficiency
random_search_class = RandomizedSearchCV(classification_pipeline, param_distributions=param_dist_class, n_iter=10, cv=TimeSeriesSplit(n_splits=5), n_jobs=-1)
random_search_reg = RandomizedSearchCV(regression_pipeline, param_distributions=param_dist_reg, n_iter=10, cv=TimeSeriesSplit(n_splits=5), n_jobs=-1)


try:
    random_search_class.fit(X_train, y_class_train)
    y_class_pred = random_search_class.best_estimator_.predict(X_test_clean)
    class_accuracy = accuracy_score(y_class_test_clean, y_class_pred)
    print(f"Classification Accuracy: {class_accuracy}")
except Exception as e:
    print(f"Classification model training failed with error: {e}")

# Regression
try:
    random_search_reg.fit(X_train, y_reg_train)
    y_reg_pred = random_search_reg.best_estimator_.predict(X_test_clean)
    reg_rmse = mean_squared_error(y_reg_test_clean, y_reg_pred, squared=False)
    print(f"Regression RMSE: {reg_rmse}")
except Exception as e:
    print(f"Regression model training failed with error: {e}")

# Error Analysis (Basic): Print out a few instances where the classification model was incorrect
errors = y_class_test_clean != y_class_pred
print("Some incorrect classifications:")
print(df.loc[errors].head())


print(f"Classification Accuracy: {class_accuracy}")
print(f"Regression RMSE: {reg_rmse}")
```
